```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
font = {        'size'   : 22}
mpl.rc('font', **font)
```

```python
df = pd.read_csv("./bases/education/EdStatsAjustado.csv")
```

```python
dfIndicatorsLabor = df[df.IndicatorCode.isin(['SE.XPD.TOTL.GD.ZS'])]
dfIndicatorsLabor = dfIndicatorsLabor[['CountryCode','IndicatorCode','Year','Value']].copy()
df2 = dfIndicatorsLabor.pivot_table(index=['CountryCode','Year'], columns='IndicatorCode', dropna=False)
df2.columns = df2.columns.droplevel().rename(None)
df2 = df2.reset_index()
df2.rename(columns={'CountryCode':'countrycode','Year':'year'},inplace=True)
df2
```

```python
dfmasc = pd.read_csv("./bases/suicide/suicidioMasculino.csv")
dfmasc = dfmasc.join(df2.set_index(['countrycode','year']), on=['countrycode','year'])
dfmasc
```

```python
dfcorr = pd.DataFrame(columns=['country','une1524', 'une2534', 'une3554','une5574','une75'])
def calculate_corr(group):
    a = group['suicidio1524'].corr(group['SL.UEM.TOTL.MA.ZS'])
    b = group['suicidio2534'].corr(group['SL.UEM.TOTL.MA.ZS'])
    c = group['suicidio3554'].corr(group['SL.UEM.TOTL.MA.ZS'])
    d = group['suicidio5574'].corr(group['SL.UEM.TOTL.MA.ZS'])
    e = group['suicidio75'].corr(group['SL.UEM.TOTL.MA.ZS'])
    
    
#dfmasc.groupby('country').apply(calculate_corr)
for name, group in dfmasc.groupby('country'):
    a = group['suicidio1524'].corr(group['SL.UEM.TOTL.MA.ZS'])
    b = group['suicidio2534'].corr(group['SL.UEM.TOTL.MA.ZS'])
    c = group['suicidio3554'].corr(group['SL.UEM.TOTL.MA.ZS'])
    d = group['suicidio5574'].corr(group['SL.UEM.TOTL.MA.ZS'])
    e = group['suicidio75'].corr(group['SL.UEM.TOTL.MA.ZS'])
    dfcorr = dfcorr.append(pd.Series([name,a,b,c,d,e], index=dfcorr.columns ),ignore_index=True)
    dfcorr.rename(columns={'une1524':'15-24','une2534':'25-34','une3554':'35-54','une5574':'55-74','une75':'75+'},inplace=True)
    #dfcorr.append(, ignore_index=True)
#dfcorr.append(, ignore_index=True)
dfcorr = dfcorr.melt(id_vars=["country"], 
        var_name="nome",
        value_name="correlacao")
dfcorr
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

bplot=sns.boxplot(y='correlacao', x='nome', 
                 data=dfcorr, 
                 width=0.5,
                 palette="colorblind")
 
# add stripplot to boxplot with Seaborn
bplot=sns.stripplot(y='correlacao', x='nome', 
                   data=dfcorr, 
                   jitter=True, 
                   marker='o', 
                   alpha=0.5,
                   color='black')
ax.grid(b=True, which='major', linewidth=1)
ax.set_xlabel('Faixa etÃ¡ria')
ax.set_ylabel('CorrelaÃ§Ã£o SuicÃ­dio x Desemprego')
ax.set_title("CorrelaÃ§Ã£o entre suicÃ­dios masculinos e taxa de desemprego")
```
